### Задача 1
При помощи библиотеки NetworkX создайте ненаправленный граф, который состоит из 5 вершин и 4 ребер. Выведите такой граф на экран.

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
Graf = nx.Graph()

nodes = ['A', 'B', 'C', 'D', 'E']

edges = [('A', 'B'), ('B', 'C'), ('B', 'D'), ('B', 'E'), ('B', 'A')]

Graf.add_nodes_from(nodes)
Graf.add_edges_from(edges)

nx.draw_networkx(Graf,
        with_labels=True, 
        node_color='g',   
        node_size=500)    
plt.show()
```

### Задача 4
При помощи библиотеки NetworkX создайте ненаправленный взвешенный полный граф, у которого 6 вершин (A,B,C,D,E,F). Веса рёбер такого графа пусть имеют значения 1,2,3,4,5 и т.д. по порядку. Выпишите кратчайшие пути между теми вершинами, которые имеют больше одного ребра на своём пути.

```python
G = nx.complete_graph(6)
eg = list(G.edges)
weights = set()
for i in range(len(eg)):
    weights.add((eg[i][0], eg[i][1], i+1))
G.add_weighted_edges_from(weights)

pos = nx.circular_layout(G)

# выбираем что будет в качестве лейблов (веса)
labels = nx.get_edge_attributes(G, 'weight')

# выводим граф на экран
nx.draw_networkx(G,
        pos,
        with_labels=True,
        node_color='lightblue',
        node_size=500)

# добавляем отрисовку лейблов для рёбер
nx.draw_networkx_edge_labels(Graf, pos, edge_labels=labels)
plt.show()
```

```python
for i in range(6):
    for j in range(6):
        if i != j:
            shortest = nx.shortest_path(G, i, j, weight='weight')
            len_shortest = nx.shortest_path_length(G, i, j, weight='weight')

            if len(shortest) == 2:
                continue
            print(f"кратчайший путь из {i} в {j}: {shortest} с длиной {len_shortest}")
```
